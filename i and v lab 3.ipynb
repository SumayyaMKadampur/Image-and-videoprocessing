```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from numpy.fft import fftshift, fft2

# Step 1: Define the size of the image and create the g1(x, y) signal
width, height = 150, 150
g1 = np.zeros((height, width), dtype=np.uint8)

# Set the right half to white (255)
g1[:, width // 2:] = 255

# Step 2: Display the 2D plot of g1(x, y)
plt.figure(figsize=(6, 6))
plt.imshow(g1, cmap='gray')
plt.title('2D Spatial Domain (g1(x, y))')
plt.colorbar()
plt.show()

# Step 3: Create a 3D plot of g1(x, y)
x = np.arange(0, width)
y = np.arange(0, height)
x, y = np.meshgrid(x, y)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, g1, cmap='gray', edgecolor='none')
ax.set_title('3D Spatial Domain (g1(x, y))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Intensity')
plt.show()

# Step 4: Compute the 2D DFT of g1(x, y) and shift zero frequency to the center
dft_g1 = fftshift(fft2(g1))

# Compute the magnitude spectrum
magnitude_spectrum = np.abs(dft_g1)

# Step 5: Display the 2D magnitude spectrum of g1(u, v)
plt.figure(figsize=(6, 6))
plt.imshow(np.log(1 + magnitude_spectrum), cmap='gray')
plt.title('2D Magnitude Spectrum (g1(u, v))')
plt.colorbar()
plt.show()

# Step 6: Create a 3D plot of the magnitude spectrum
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Adjust the x, y meshgrid for the frequency domain
u = np.fft.fftfreq(width).reshape(-1, 1)
v = np.fft.fftfreq(height).reshape(-1, 1)
U, V = np.meshgrid(u, v)

# Plot the 3D surface with better coloring and visibility
ax.plot_surface(U, V, np.log(1 + magnitude_spectrum), cmap='inferno', edgecolor='none')
ax.set_title('3D Magnitude Spectrum (g1(u, v))')
ax.set_xlabel('Frequency X')
ax.set_ylabel('Frequency Y')
ax.set_zlabel('Magnitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from PIL import Image
from numpy.fft import fftshift, fft2

# Step 1: Define the size of the image and create the g2(x, y) signal (2D impulse signal)
width, height = 150, 150
impulse_size = 10
g2 = np.zeros((height, width), dtype=np.uint8)

# Calculate the center position of the image
center_x, center_y = width // 2, height // 2

# Calculate the starting and ending positions of the 10x10 impulse block
start_x, start_y = center_x - impulse_size // 2, center_y - impulse_size // 2

# Set the 10x10 block in the center to white (255)
g2[start_y:start_y + impulse_size, start_x:start_x + impulse_size] = 255

# Display the 2D plot of g2(x, y)
plt.figure(figsize=(6, 6))
plt.imshow(g2, cmap='gray')
plt.title('2D Spatial Domain (g2(x, y))')
plt.colorbar()
plt.show()

# Create a 3D plot of g2(x, y)
x = np.arange(0, width)
y = np.arange(0, height)
x, y = np.meshgrid(x, y)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, g2, cmap='gray')
ax.set_title('3D Spatial Domain (g2(x, y))')
plt.show()

# Step 2: Compute the 2D DFT of g2(x, y) and shift zero frequency to the center
dft_g2 = fftshift(fft2(g2))

# Compute the magnitude spectrum
magnitude_spectrum = np.abs(dft_g2)

# Step 3: Display the 2D magnitude spectrum of g2(u, v)
plt.figure(figsize=(6, 6))
plt.imshow(np.log(1 + magnitude_spectrum), cmap='gray')
plt.title('2D Magnitude Spectrum (g2(u, v))')
plt.colorbar()
plt.show()

# Create a 3D plot of the magnitude spectrum
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, np.log(1 + magnitude_spectrum), cmap='inferno')
ax.set_title('3D Magnitude Spectrum (g2(u, v))')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from PIL import Image
from numpy.fft import fftshift, fft2

# Step 1: Load the grayscale image
image_path = 'images.jpeg'  # Replace with the actual path to your image
image = Image.open(r'C:\Users\Lenovo\Pictures\images.jpeg').convert('L')
image_data = np.array(image)

# Step 2: Display the 2D plot of the image
plt.figure(figsize=(6, 6))
plt.imshow(image_data, cmap='gray')
plt.title('2D Spatial Domain Image')
plt.colorbar()
plt.show()

# Step 3: Display the 3D plot of the image in the spatial domain
x = np.arange(0, image_data.shape[1])
y = np.arange(0, image_data.shape[0])
x, y = np.meshgrid(x, y)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, image_data, cmap='gray', edgecolor='none')
ax.set_title('3D Spatial Domain Image')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Intensity')
plt.show()

# Step 4: Compute the 2D DFT of the image and shift the zero frequency component to the center
dft_image = fftshift(fft2(image_data))

# Compute the magnitude spectrum
magnitude_spectrum = np.abs(dft_image)

# Step 5a: Display the 2D magnitude spectrum
plt.figure(figsize=(6, 6))
plt.imshow(np.log(1 + magnitude_spectrum), cmap='gray')  # Log scaling for better visibility
plt.title('2D Magnitude Spectrum')
plt.colorbar()
plt.show()

# Step 5b: Display the 3D plot of the magnitude spectrum
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Adjust the x, y meshgrid for the frequency domain
u = np.fft.fftfreq(image_data.shape[1])
v = np.fft.fftfreq(image_data.shape[0])
U, V = np.meshgrid(u, v)

ax.plot_surface(U, V, np.log(1 + magnitude_spectrum), cmap='inferno', edgecolor='none')
ax.set_title('3D Magnitude Spectrum')
ax.set_xlabel('Frequency X')
ax.set_ylabel('Frequency Y')
ax.set_zlabel('Magnitude')
plt.show()
```
