```python
#Implementation of  All Edge Detection techniques.

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread(r"C:\Users\Lenovo\Pictures\rajasthan.jpg", cv2.IMREAD_GRAYSCALE)

# Function to display multiple images
def display_images(titles, images):
    plt.figure(figsize=(10, 7))
    for i in range(len(images)):
        plt.subplot(2, 3, i+1), plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.xticks([]), plt.yticks([])
    plt.show()

# 1. Sobel Edge Detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Horizontal
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Vertical
sobel_combined = cv2.magnitude(sobel_x, sobel_y)  # Magnitude of gradients

# 2. Prewitt Edge Detection (Using a custom kernel, as OpenCV doesn't provide it)
prewitt_kernel_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
prewitt_kernel_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
prewitt_x = cv2.filter2D(image, cv2.CV_64F, prewitt_kernel_x)  # Convert to float32
prewitt_y = cv2.filter2D(image, cv2.CV_64F, prewitt_kernel_y)  # Convert to float32
prewitt_combined = cv2.magnitude(prewitt_x, prewitt_y)

# 3. Roberts Edge Detection (Using custom kernels)
roberts_kernel_x = np.array([[1, 0], [0, -1]])
roberts_kernel_y = np.array([[0, 1], [-1, 0]])
roberts_x = cv2.filter2D(image, cv2.CV_64F, roberts_kernel_x)
roberts_y = cv2.filter2D(image, cv2.CV_64F, roberts_kernel_y)
roberts_combined = cv2.magnitude(roberts_x, roberts_y)

# 4. Laplacian Edge Detection
laplacian = cv2.Laplacian(image, cv2.CV_64F)

# 5. Canny Edge Detection
canny_edges = cv2.Canny(image, 100, 200)

# Display the results
titles = ['Original Image', 'Sobel Combined', 'Prewitt Combined', 'Roberts Combined', 'Laplacian', 'Canny']
images = [image, sobel_combined, prewitt_combined, roberts_combined, laplacian, canny_edges]
display_images(titles, images)    
```
