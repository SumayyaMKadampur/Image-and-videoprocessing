```python
#Implementation of linear transformation.
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r'C:\Users\Lenovo\Pictures\gujarat.jpg')

# Convert the image to grayscale (if not already)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define transformation parameters
a = 1.5  # Scaling factor
b = 50   # Offset

# Apply the linear transformation
transformed_image = np.clip(a * gray_image + b, 0, 255).astype(np.uint8)

# Concatenate the images side by side
combined_image = np.hstack((gray_image, transformed_image))

# Display the images using matplotlib
plt.figure(figsize=(12, 6))
plt.imshow(combined_image, cmap='gray')
plt.title('Original and Transformed Image')
plt.axis('off')  # Hide axis
plt.show()
```

```python
#Implementation of Contrast stretching.

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image = cv2.imread(r"C:\Users\Lenovo\Downloads\camera.webp", cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded properly
if image is None:
    raise ValueError("Image not found. Please check the file path.")

# 1. Negative Transformation
negative_image = 255 - image

# 2. Contrast Stretching
min_val, max_val = np.min(image), np.max(image)
contrast_stretched_image = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)

# 3. Thresholding
_, thresholded_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

# 4. Gamma Correction
gamma = 2.0  # Change this value to experiment with different gamma levels
gamma_corrected_image = np.array(255 * (image / 255) ** gamma, dtype='uint8')

# Display the original and transformed images
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(2, 3, 2)
plt.title("Negative Transformation")
plt.imshow(negative_image, cmap='gray')

plt.subplot(2, 3, 3)
plt.title("Contrast Stretching")
plt.imshow(contrast_stretched_image, cmap='gray')

plt.subplot(2, 3, 4)
plt.title("Thresholding")
plt.imshow(thresholded_image, cmap='gray')

plt.subplot(2, 3, 5)
plt.title("Gamma Correction")
plt.imshow(gamma_corrected_image, cmap='gray')

plt.tight_layout()
plt.show()
    
```

```python
#Implementation of  gray scale slicing.
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r'C:\Users\Lenovo\Downloads\camera.webp')

# Convert the image to grayscale (if not already)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define the gray level range to slice
low_threshold = 100  # Lower bound of the gray level range
high_threshold = 150 # Upper bound of the gray level range

# Create the gray level sliced image
sliced_image = np.where((gray_image >= low_threshold) & (gray_image <= high_threshold), 255, 0).astype(np.uint8)

# Save the sliced image
cv2.imwrite('gray_level_sliced_image.jpg', sliced_image)

# Display the original and sliced images side by side
combined_image = np.hstack((gray_image, sliced_image))

# Display the images
plt.figure(figsize=(12, 6))
plt.imshow(combined_image, cmap='gray')
plt.title('Original and Gray Level Sliced Image')
plt.axis('off')  # Hide axis
plt.show()

    
```

```python
#Implementation of histogram to a image.
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale mode
image = cv2.imread(r'C:\Users\Lenovo\Downloads\camera.webp', 0)

# Calculate the histogram
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# Plot the image and its histogram
plt.figure(figsize=(12, 5))

# Plot the original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Plot the histogram
plt.subplot(1, 2, 2)
plt.plot(histogram, color='black')
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])

plt.show()

    
```

```python
#Implementation of histogran  before and after equalization 

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread(r'C:\Users\Lenovo\Downloads\camera.webp', 0)

# Calculate histogram before equalization
hist_before = cv2.calcHist([image], [0], None, [256], [0, 256])

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Calculate histogram after equalization
hist_after = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

# Plot the original image, equalized image, and their histograms
plt.figure(figsize=(12, 8))

# Plot original image
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Plot histogram before equalization
plt.subplot(2, 2, 2)
plt.plot(hist_before, color='black')
plt.title('Histogram Before Equalization')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])

# Plot equalized image
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

# Plot histogram after equalization
plt.subplot(2, 2, 4)
plt.plot(hist_after, color='black')
plt.title('Histogram After Equalization')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()

    
```
